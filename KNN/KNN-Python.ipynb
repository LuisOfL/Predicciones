# KNN-K-Nearest Neighbor (Python)

Funciones

```python
def DisEuc(p1, p2):
    return (((p1[0]-p2[0])**2)+((p1[1]-p2[1])**2))**0.5

def MinDis(p, X, y):
    l = []
    for xi, yi in zip(X, y):
        n = DisEuc(p, xi)
        l.append([n, yi])
    return sorted(l, key=lambda x: x[0])

def KNN(p, X, y, k):
    vecinos = MinDis(p, X, y)[:k]
    dic = {}
    for distancia, etiqueta in vecinos:
        if etiqueta in dic:
            dic[etiqueta] += 1
        else:
            dic[etiqueta] = 1
    return max(dic, key=dic.get)
```

Ejemplo
