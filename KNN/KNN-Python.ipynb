# KNN-K-Nearest Neighbor (Python)

Funciones

```python
def DisEuc(p1, p2):
    return (((p1[0]-p2[0])**2)+((p1[1]-p2[1])**2))**0.5

def MinDis(p, X, y):
    l = []
    for xi, yi in zip(X, y):
        n = DisEuc(p, xi)
        l.append([n, yi])
    return sorted(l, key=lambda x: x[0])

def KNN(p, X, y, k):
    vecinos = MinDis(p, X, y)[:k]
    dic = {}
    for distancia, etiqueta in vecinos:
        if etiqueta in dic:
            dic[etiqueta] += 1
        else:
            dic[etiqueta] = 1
    return max(dic, key=dic.get)
```

Ejemplo 

```python
import random


x = []
y = []

for _ in range(50):
    peso = random.randint(140, 180)   
    cantidad = random.randint(5, 10)  
    x.append([peso, cantidad])
    y.append("Manzana")

# Generar 50 naranjas
for _ in range(50):
    peso = random.randint(180, 220)   
    cantidad = random.randint(3, 7)   
    x.append([peso, cantidad])
    y.append("Naranja")
```

Graficacion

```python
import matplotlib.pyplot as plt
for point,label in zip(x,y):
    if label == 'Manzana':
        plt.scatter(point[0],point[1],color='red')
    elif label == 'Naranja':
        plt.scatter(point[0],point[1],color='orange')
plt.show()
```

Prediccion

```python
nuevo = [175, 6]
n = KNN(nuevo,x,y, 10)
print(n)
```
